```python
import fitz  # PyMuPDF
import pandas as pd
from sklearn.model_selection import train_test_split

# Open the PDF file
file_path = '/mnt/data/toaz.info-4256442-sylvia-plath-the-collected-poems-pr_74a7f929f68221792d41a5817d18bcca.pdf'
pdf_document = fitz.open(file_path)

# Extract text from each page after page 19
text_data = []
for page_num in range(19, pdf_document.page_count):
    page = pdf_document.load_page(page_num)
    text = page.get_text()
    text_data.append(text)

# Combine all the text into one string
combined_text = "\n".join(text_data)

# Split the combined text into smaller segments (assuming each segment is separated by double newlines)
text_segments = combined_text.split("\n\n")

# Splitting the segments into training and testing sets (80% training, 20% testing)
train_segments, test_segments = train_test_split(text_segments, test_size=0.2, random_state=42)

# Create a DataFrame for each set
train_df = pd.DataFrame({'text': train_segments, 'type': 'train'})
test_df = pd.DataFrame({'text': test_segments, 'type': 'test'})

# Combine both DataFrames
#all_data_df = pd.concat([train_df, test_df])

# Save the data to a CSV file
#csv_file_path = '/mnt/data/sylvia_plath_collected_poems.csv'
#all_data_df.to_csv(csv_file_path, index=False)
```
